import the required packages

```python
import pandas as pd
import numpy as np
import io
import itertools
import math
import matplotlib.pyplot as plt
import pyam
```

```python
df = pyam.IamDataFrame(data='data/ar6_snapshot_1682256076.xlsx')
```

```python
df.model
remind_df= df.filter(model='REMIND 2.1')
```

```python
remind_df.scenario
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter().plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

```python
gem_df=df.filter(model="GEM-E3_V2021")
fig, ax = plt.subplots(figsize=(8, 8))
gem_df.filter().plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```
