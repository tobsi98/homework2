This work shows the final energy consumption in the European area in different sceanrios. Furthermore the development of usage of coal and solar is analyzed.
The model REMIND 2.1 is used and different scenarios are shown in the graphs.

import the required packages

```python
import matplotlib.pyplot as plt
import pyam
```

First the ar6_snapshot from the ar6 scenario explorer is imported in an pyam dataframe

```python
df = pyam.IamDataFrame(data='data/ar6_snapshot_1682257715.xlsx')
remind_df= df.filter(model='REMIND 2.1')
remind_df.scenario
```

Using the REMIND 2.1 Model, which is the abbreviation REgional Model of INvestments and Development

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter(variable="Final Energy").plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

In this plot the Final Energy consumption is plotted. Until the year 2040 in all scenarios the final energy consumption is decreasing. After the year 2040 in the senario R2p1_SSP5-PkBudg900 (High Growth Scenario) which is characterized by fossil-fuel led economic growth, the energy consumption is rising. The four "CEMICS" (climate change mitigation scenarios) sceanrios are very close to each other. Alle CEMICS scenarios have the goal to show a way how to reach the 1p5 degree climate goal. (paris agreement). Only two scenarios are showing a further declining energy consumption until the year 2100

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter(variable="Final Energy|Electricity").plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

In this figure the electricity consumption is shown. In all scenarios the use of electricity is increasing dramatically after the year 2025.

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter(variable="Final Energy|Gases").plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

In this figure the electricity the usage of natural gas for the final energy consumption is shown. As we can see, the usage of gas hast to be reduced to nearly 1/6 until the year 2050 to reach the 1p5 degree goal.

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter(variable="Secondary Energy|Electricity|Coal").plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

In this figure the usage of coal for electricity production is shown. In mostly all scenarios we have to reduce the usage of coal for electricity to zero around the year 2030.

```python
fig, ax = plt.subplots(figsize=(6, 6))
remind_df.filter(variable="Secondary Energy|Electricity|Solar").plot(ax=ax,legend=True)
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 0.8))
ax.grid()
```

In this figure the electricity production of solar energy is shown. All scenarios give us a nearly exponential growth path until 2040 for the solar energy usage. After 2040 some scenarios are showing a saturation. Other scenarios are changing from an exponential growth to a linear growth after 2040.
